Handling imbalance dataset
Imbalanced data refers to those types of datasets where the target class has an uneven distribution of observations, i.e one class label has a very high number of observations and the other has a very low number of observations. 

Model will ge biased towards higest no of observations.

Resampling (Oversampling and Undersampling) - Upsampling and Downsampling with the help of this techniques imbalance data can be handeled.

Upsamping: We try to increase the data points at low no of observations.

Downdampling: We try to decrease the data poionts at hign no of observations.


```python
import numpy as np
import pandas as pd

# set the random seed for reproductivity
# The numpy random seed is a numerical value that generates a new set or repeats pseudo-random numbers. The value in the numpy random seed saves the state of randomness. 
# If we call the seed function using value 1 multiple times, the computer displays the same random numbers.
# A random seed is used to ensure that results are reproducible. In other words, using this parameter makes sure that anyone who re-runs your code will get the exact same outputs.
# Reproducibility is an extremely important concept in data science and other fields.

np.random.seed()

# Create dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
# Creating dataframe with imbalanced dataset

# Generates random samples from a normal (Gaussian) distribution. When you standardize a normal distribution, the mean becomes 0 and the standard deviation becomes 1
# here in below function loc is mean and scale is std with size as no of observations we want in dataset
class_0 = pd.DataFrame ({
    'feature_1' : np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2' : np.random.normal(loc=0, scale=1, size=n_class_0),
    'target' : [0] * n_class_0 # 0 will get multiplied by 900 data points that means 900, 0s will get created
})


class_1 = pd.DataFrame ({
    'feature_1' : np.random.normal(loc=0, scale=1, size=n_class_1),
    'feature_2' : np.random.normal(loc=0, scale=1, size=n_class_1),
    'target' : [1] * n_class_1 # 0 will get multiplied by 900 data points that means 900, 0s will get created
})
```

```python
class_0.head()
```

```python
class_1.head()
```

```python
# concat two dataframes, reset_index : Do not try to insert index into dataframe columns. This resets the index to the default integer index.
df = pd.concat([class_0,class_1]).reset_index(drop=True) 
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts() # displays count of 0s and 1s
```

```python
## Upsampling to increase 1s data points 

df_minority = df[df['target']==1] # it will return all 1, this dataframe will be used to upsample 1s
df_majority = df[df['target']==0] # it will return all 0, this datafrma can be used to downsample 0s
```

```python
df_minority
```

```python
df_majority
```

```python
from sklearn.utils import resample # this will extrapolate, it will create extra points and try to match with majority points

df_minority_upsampled = resample(df_minority, replace=True, # this will resample with replacements
         n_samples = len(df_majority),
         random_state = 42 #https://www.linkedin.com/pulse/story-behind-random-state-42-kriti-biswas/
         )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_minority_upsampled['target'].value_counts()
```

```python
# now combine both dataframes to get balanced final data set which will have 900 0s and 900 1s 
df_upsampled = pd.concat([df_majority,df_minority_upsampled])
```
