Web Scrapping:

Web scraping is an automatic method to obtain large amounts of data from websites.

Most of this data is unstructured data in an HTML format which is then converted into structured data in a spreadsheet or a database so that it can be used in various applications.

We may need to acquire some data from web.

```python
# How to do a scrapping
# Flipkart product reviews - Objective is to scrap all these reviews for sentiment analysis about product.


import requests #it is an http libraby written in python
from bs4 import BeautifulSoup as bs #Beautiful Soup is a Python library for pulling data out of HTML and XML files.
# It works with your favorite parser to provide idiomatic ways of navigating, searching, and modifying the parse tree. It commonly saves programmers hours or days of work.

from urllib.request import urlopen  #Urllib package is the URL handling module for python. It is used to fetch URLs (Uniform Resource Locators). 
# It uses the urlopen function and is able to fetch URLs using a variety of different protocols. 
# Urllib is a package that collects several modules for working with URLs, such as: urllib

import logging #The logging module in Python is a ready-to-use and powerful module that is designed to meet the needs of beginners as well as enterprise teams.
# It is used by most of the third-party Python libraries, so you can integrate your log messages with the ones from those libraries to produce a homogeneous log for your application.

```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "iphone14pro" #creating URL - get query search part from flipkart search url
```

```python
flipkart_url
```

```python
urlopen(flipkart_url) # urlopen method is used to open the url automatically
url_client = urlopen(flipkart_url)
```

```python
flipkart_page = url_client.read() # this will display all the data on url page which was in search (iphone14pro) in the form of html source code of the page as scrap - all the related search.
```

```python
flipkart_page #dump data, unstructured, unreadable - to convert this kind of data into some kind of structured and some what readable format - beautifulsoap comes into a picture
```

```python
flipkart_html = bs(flipkart_page, 'html.parser') #A parser is a program that is part of the compiler, and parsing is part of the compiling process.
# Parsing happens during the analysis stage of compilation. In parsing, code is taken from the preprocessor, broken into smaller pieces and 
# analyzed so other software can understand it
#bs will beautify the html code normally
```

```python
# Our objective is to get in to all related product details (URLs) one by one which are available on search page. 
# Our code need to open these each product URLs 

# Lets inspect this webpage - goto that flipkart page and press F12 on keyboard - you will get html code and flipkart page side by side and once you click on inspect button on 
# top left corner of the html source code page, and move cursur to flipkart page - it will highlight the boxes on flipkart page and relevant code on html source code page.

#we need to catch each product's URL (product detail page URL) which are present in html source code page e.g. 

# Product detail page URL: apple-iphone-14-pro-space-black-128-gb/p/itm6e6711f2982c8?pid=MOBGHWFHZBHSYBBW&lid=LSTMOBGHWFHZBHSYBBWR0WJJK&marketplace=FLIPKART&q=iphone+14+pro&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&fm=organic&iid=4dbae728-a6e8-4b97-b1ad-40f6b71ca5c0.MOBGHWFHZBHSYBBW.SEARCH&ppt=hp&ppn=homepage&ssid=it0gyst9cg0000001677078245040&qH=73a41d19c3188cc2

# and our requirement is to get such URLs in a automated way

```

```python
# On flipkart product search result page we have all products and each product bigbox is represented by <div class="_1AtVbE col-12-12" > in html source code
# all of these kind of div classes (_1AtVbE) holding a dataset for each product bigbox and we try to fetch all of these big boxes (classes), class names are same and under this class name there is data for each product on search page
# flipkart_html variable has all the dump data stored so from this dump data we will try to fetch all div classes with class IDs as mentioned below
bigbox = flipkart_html.find_all("div", {"class":"_1AtVbE col-12-12"}) #all div and class where id is as mentioned here, passing this in a dictionary

#this is returning a data in a list
```

```python
len(bigbox) #this will show how many no of bigboxes (no of products on that page)
```

```python
del bigbox[0:3] #this is because first 3 boxes are of heading on page and not a product
```

```python
# now from these big boxes we need to fetch URL for each product details page i.e. html keywori 'href' inside these div classed heirarchi  - count the no of div till href keyword 
## so that we can reach out to 'href' keyword and fetch product details page URL where product reviews are available. div of div of div of div and under html tag <a> we can get URL

productlink = "https://www.flipkart.com" + bigbox[0].div.div.div.a['href'] #trying to fetch URL for 1st bigbox with using index 0 and out of this tag <a> we have to extract href URL - passing href so that we can get URL whici is availabe next to href

#append the output of href (URL) to flipkart.com so that we can able to get to the product detail page
#this is for single bigbox, need to do this with all the bigboxes with for loop as in next code line...
```

```python
for i in bigbox:
    print("flipkar.com" + i.div.div.div.a['href']) #below are the URLs for each product details page
```

```python
productlink
```

```python
# now we will open each URL for the product details page and we need to extract all data related to product reviews
product_req = requests.get(productlink) #single URL as inpurt but we can provide multiple URLs using for loop
```

```python
product_req.text #with .text we can able to see scrap of the page - html source code
```

```python
#since this is not in human readable format, lets beautify it little bit with the help of beautiful soap
product_html = bs(product_req.text, 'html.parser') #out of this html page we have to extract only reviews and ratings
```

```python
# now go to that product page and press F12, with this we are able to inspect the page and again try to find out big boxes where each comment and ratings are updated
comment_box = product_html.find_all("div",{"class":"_16PBlm"}) #this will show all classes with name _16PBlm where comments are updated

```
