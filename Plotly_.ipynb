```python
#plotly can generate interactive plots
import plotly.graph_objects as go # user can interact with graph, if you hover mounse pointer in data points on graph it will show info.
```

```python
fig = go.Figure() # it creates canvas for visual
```

```python
fig.add_trace(go.Scatter(x = [1,2,3,4], y = [5,6,7,8], mode='markers')) #add trace will add graph to canvas as per parameters given, mode = type of visual to display datapoits 
#e.g. lines or markers (symbols)
```

```python
fig.add_trace(go.Scatter(x = [1,2,3,4], y = [5,6,7,8], mode='lines'))
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = ' markers',marker_size = 10*tips['size'])) #marker size will show size col in data set, we have multiplied it with 10 just to displya bigger size
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4], y = [5,6,7,8]))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x = [1,2,3,4]))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x = tips.total_bill))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = tips.total_bill, y = tips.tip, mode = ' markers',z = 10*tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = tips.total_bill, y = tips.tip, mode = ' lines',z = 10*tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1,2,3,4], y = [2,6,1,8], mode = ' lines',z = [3,7,1,8]))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1,2,3,4], y = [2,6,1,8], mode = ' markers', z = [3,7,1,8]))
```
